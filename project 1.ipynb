```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#Task 1: Getting the Dataset
df = pd.read_csv('data.csv')
df
```

```python
#Task 2: Basic Visualization
x = df["Degree of endangerment"].unique()
h = pd.value_counts(df["Degree of endangerment"])
plt.bar(x,h,width = 0.9,align = 'center',color = ["pink","brown", "green","blue","purple"])
plt.xticks(rotation = 40)

plt.xlabel("Degree of Enadangerment")
plt.ylabel("Number of languages")
plt.title("Bar Chart")

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='gray')

plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
#Task 3: East African Languages
list = ["Kenya", "Ethiopia","Uganda", "Sudan","Somalia", "Rwanda","Burundi"]
sv = df[df.Countries.isin(list)]
H = pd.value_counts(sv["Degree of endangerment"])

X = df["Degree of endangerment"].unique()
plt.bar(X,H, width = 0.9,align = 'center',color = ["pink","brown", "green","blue","purple"])
plt.xticks(rotation = 40)

plt.xlabel("Degree of Enadangerment")
plt.ylabel("Number of languages")
plt.title("Bar Chart")

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='gray')

plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
